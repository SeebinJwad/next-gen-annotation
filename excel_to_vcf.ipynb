```python
# sub_name must be p-v or v-p
def create_vcf(sub_name):
    df = pd.read_excel(file_name, sheet_name=sub_name + " " + variant)
    vcf_file = parent + "_" + variant + "_" + sub_name + ".vcf"
    with open(vcf_file, 'w') as vcf:
        vcf.write(header)
    df.to_csv(vcf, sep="\t", mode='a', index=False)
    return None
```

```python
for x in range(3):
    for y in range(4):
        
        parent = ids[x*5]
        variant = ids[x*5 + y+1]
        
        file_name = "results/VE_" + parent + "_VE_" + variant + ".xlsx"

        pv_df = pd.read_excel(file_name, sheet_name="p-v " + variant)      
        vp_df = pd.read_excel(file_name, sheet_name="v-p " + variant)

        pv_vcf = parent + "_" + variant + "_p-v.vcf"
        vp_vcf = parent + "_" + variant + "_v-p.vcf"
        
        with open(pv_vcf, 'w') as vcf:
            vcf.write(header)
        
        with open(vp_vcf, 'w') as vcf:
            vcf.write(header)
        
        pv_df.to_csv(pv_vcf, sep="\t", mode='a', index=False)
        vp_df.to_csv(vp_vcf, sep="\t", mode='a', index=False)
```
